```python
import numpy as np
import matplotlib.pyplot as plt

def newton_raphson(f, f_prime, x0, max_iter=100, tol=1e-7):
    root = x0
    history = [x0]

    for i in range(max_iter):
        delta = f(root) / f_prime(root)
        root -= delta
        history.append(root)

        if abs(delta) < tol:
            break

    return root, history

#defining the function and its derivative
f = lambda x: 4*x**3 - 5*x +21
f_prime = lambda x: 12*x**2 -5

#initial guess for the root
x0 = -1.5

#applying the Newton-Raphson method
root, history = newton_raphson(f, f_prime, x0)

print("Estimated root:", root)

plt.plot(history)
plt.xlabel("Iteration")
plt.ylabel("Root estimate")
plt.title("Convergence of Newton-Raphson Method")
plt.show()
```
